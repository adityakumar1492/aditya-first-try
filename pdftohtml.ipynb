```python
all_pages = []
with open('NLP_compressed.pdf', "+rb") as pdf:
    reader = PdfReader(pdf)
    total_pages = len(reader.pages)

    for n in range(total_pages):
        page = reader.pages[n]
        data = page.extract_text()
        if data!='':
            html_data = '''<div class="step">'''+data+'''</div>'''
            all_pages.append(html_data)

        
```
